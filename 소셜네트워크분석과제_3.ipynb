# 그래프의 기본 개념

## 그래프 이론
- 객체들 간의 네트워크 관계 모델을 수학적으로 표현된 그래프로 나타내는
연구 분야 그래프는 노드(Node, vertex)로 구성, 이러한 노드들은 링크(Link, edge)로 연결되어 있음.


## 무방향 네트워크

로잉 네트워크 등

## 이진 네트워크
- 관계의 유무만을 표현

## 가중 네트워크
- 관계의 유무 + 관계의 정도

# 그래프의 표현 방법  (파이썬으로 작성한 과제에서 실습함.)

## 인접행렬
- 인접한 노드들 간의 연결 관계를 행렬(Matrix)로 표현, 네트워크 상에서 노드의 개수가 n인 경우 인접 행렬은 nxn의 정방 행렬로 표현
- 행과 열의 이름은 노드를 나타내며, 행과 열이 교차하는 지점의 행렬 값은 관계(link)의 유무를 1과 0으로 나타냄. 행의 이름은 링크가 출발하는 노드를 나타내고 열의 이름은 링크가 도착하는 노드를 나타낸다. 그림 2-21>은 방향 네트워크의 인접행렬로 나타낸 방식을 보여주고 있음.
- 무방향 네트워크의 경우 모든 관계가 쌍방향이어서 방향성이 상쇄되는 경우와 동일하게 됨. 즉 인접행렬은 대칭적인 특징을 가짐



## 인접리스트
- 그래프의 각 정점마다 해당 정점에서 나가는 간선의 목록을 저장해서 그래프로 표현
- 각 정점마다 하나의 연결리스트를 갖는 방식으로 구현




## 노드리스트
- 인접한 링크가 있는 노드만을 표기하는 방법으로 노드들 만의 목록
- 아래의 그림은 은 예시 그래프를 대상으로 노드 리스트를 작성한 것임.
- 맨 첫번째 열은 그래프에 존재하는 노드들의 목록을 나타낸다. 그 우측열 부터는 해당되는 그 노드에 연결된 노드의 목록을 나타낸다. 여기서 전체 행의 개수는 전체 노드의 개수와 일치하며, 노드별로 연결된 링크의 개수가 많을수록 열의 개수도 증가하게 된다. 방향성의 표시를 위해 incoming, outgoing을 구분한다

## 분석 대상에 따른 분류
- 전체 네트워크 vs. 하위 네트워크
    - 전체 네트워크 : 네트워크를 구성하고 있는 모든 노드를 포함한 네트워크를 말함. 완전 네트워크(whole network)라고도 함.
    - 하위 네트워크 : 전체 네트워크의 일부분으로 부속되어 있는 것으로 컴포넌트, 파당, 클러스터 등이 해당.

- 파당
 일반적으로 양자 관계를 제외하고 3개 이상의 노드를 포함한 것을 파당이라고 함.


- 클러스터 : 유사한 속성을 가지고 있는 노드들의 집합. 유사성에 따라 클러스터를 분류해 내는 작업을 클러스터링이라고 함.


- 에고 네트워크
    - 네트워크 상에에서 분석하고자 하는 특정 노드를 중심으로 두고, 그 노드와 직접적으로 연결된 노드들 간의 관계를 네트워크로 표현한 것. 자아중심 네트워크라고도 함.
    

# 소셜 네트워크 분석 지표 (중심성 분석)


## 중심성 지표
- 중심성 지표는 소셜 네트워크 분석 지표 중에서 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표
- 중심성 분석을 통해 전체 네트워크에서 중요한 역할을 하는 노드가 무엇인지 파악 가능
- 연결정도 중심성(Degree centrality): 한 노드가 얼마나 많은 다른 노드들과 연결관계를 맺고 있는 지 측정
- 근접 중심성(Closeness centrality): 한 노드가 얼마나 전체 네트워크의 중심에 근접해 있는지를 측정
- 매개 중심성(Betweenness centrality): 한 노드가 다른 노드들 간의 네트워크 관계 형성에 있어서 중개자 혹은 매개자 역할을 얼마나 수행하는 지 측정하는 지표
- 아이겐 벡터 중심성(Eigenvector centrality): 연결된 다른 노드들의 중요도를 함께 반영한 것
